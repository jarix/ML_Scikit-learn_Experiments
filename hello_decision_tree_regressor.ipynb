# Hello Decision Tree Regressor

```python
# Prerequisites
import sys
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
print("Python Version: ", sys.version)
print("Numpy Version: ", np.__version__)
np.set_printoptions(precision=3, suppress=True)
```

### Build a simple Decision Tree Regressor

Generate some quadratic data

```python
np.random.seed(42)

X = np.random.rand(200, 1) - 0.5
y = X**2 + 0.025 * np.random.rand(200,1)
```

plot data

```python
plt.figure(figsize=(8,5))
plt.scatter(X, y)
plt.show()
```

Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor

#tree_reg = DecisionTreeRegressor(random_state=42, criterion="squared_error")
tree_reg = DecisionTreeRegressor(max_depth=2, random_state=42, criterion="squared_error")
tree_reg.fit(X, y)
```

### Visualize Tree

```python
from sklearn.tree import export_graphviz
from graphviz import Source

export_graphviz(tree_reg, out_file="images/tree_regressor.dot", feature_names=["x1"], rounded=True, filled=True)

Source.from_file("images/tree_regressor.dot")
```
